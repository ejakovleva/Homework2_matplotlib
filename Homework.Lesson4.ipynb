### Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.array(np.cos(t))
f
```

```python
title_dict = {'fontsize': 15, 'fontweight': 'bold', 'color': 'grey', 'family': 'Calibri'}
label_dict = {'fontsize': 12, 'color': 'black', 'family': 'Calibri'}
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.plot(t, f, color='green')

plt.title('График f(t)', fontdict=title_dict)

plt.xlabel('Значения t', fontdict=label_dict)
plt.ylabel('Значения f', fontdict=label_dict)

plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')


fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3, wspace=0.3)

plt.show()
```

### Задание 4

```python
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = './creditcard.csv' 
df = pd.read_csv(DATASET_PATH, sep=',')
df.head()
```

```python
round(df.describe(),10)
```

```python
df1 = df['Class'].value_counts()
df1
```

```python
df1.plot(kind='bar', logy=True, grid=False, color='grey')

label_dict = {'fontsize': 12, 'color': 'black', 'family': 'Calibri'}
plt.xlabel('Class', fontdict=label_dict)
plt.xticks(rotation=0)
plt.set_facecolor('w')

fig.set_size_inches(8, 6)
plt.grid(axis='y', color='ghostwhite')
```

```python
df.loc[df['Class'] == 1, 'Class'].value_counts()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax.flatten()

ax1.hist(df.loc[df['Class'] == 1, 'V1'], density=True, bins=20, alpha=0.5, color='red')
ax1.grid(color='ghostwhite')
ax1.set_facecolor('w')
ax1.set_title('Class 1', color='red', loc='left', fontsize=12)


ax2.hist(df.loc[df['Class'] == 0, 'V1'], density=True, bins=20, alpha=0.5, color='grey')
ax2.grid(color='ghostwhite')
ax2.set_facecolor('w')
ax2.set_title('Class 0', color='grey', loc='left', fontsize=12)

fig.set_size_inches(12, 3)
plt.subplots_adjust(wspace=0.2)


plt.show()
```
